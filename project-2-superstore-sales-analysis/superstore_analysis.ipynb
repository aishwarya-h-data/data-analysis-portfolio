# 🛒 Superstore Sales Analysis

## Problem Statement
Analyze 4 years of retail sales data to identify revenue trends,
top-performing products, customer segments, and regional performance.

## Dataset
- **Source:** Sample Superstore Dataset (Kaggle)
- **Rows:** 9,994 orders
- **Columns:** 21 features
- **Period:** 2014 – 2017

## Tools Used
Python | Pandas | NumPy | Matplotlib | Seaborn | MySQL | Power BI

## Workflow
1. Data Loading & Understanding
2. Data Cleaning
3. Exploratory Data Analysis (EDA)
4. SQL Analysis
5. Power BI Dashboard
6. Key Insights & Conclusions

```python
# ================================================
# CELL 2: Import Libraries
# ================================================

import pandas as pd          # data manipulation
import numpy as np           # numerical operations
import matplotlib.pyplot as plt  # plotting
import seaborn as sns        # advanced visualizations

# Show plots inside notebook
%matplotlib inline

# Set consistent style for all plots
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

print("✅ Libraries imported successfully")
```

```python
# ================================================
# CELL 3: Load Dataset
# ================================================

# Load the superstore CSV file
df = pd.read_csv("superstore.csv", encoding='latin1')

# Quick look at the data
print("Shape of dataset:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

## 🔍 Section 1: Data Understanding
Getting familiar with the dataset —
its shape, columns, data types, and basic statistics.

```python
# ================================================
# CELL 5: Understand Data Structure
# ================================================

print("Dataset Info:")
print("-" * 40)
df.info()

print("\nColumn Names:")
print(df.columns.tolist())
```

```python
# ================================================
# CELL 6: Basic Statistics
# ================================================

print("Statistical Summary:")
print("-" * 40)
df.describe()
```

## 🧹 Section 2: Data Cleaning
Checking for missing values, fixing data types,
and preparing data for analysis.

```python
# ================================================
# CELL 8: Check Missing Values & Data Types
# ================================================

print("Missing Values:")
print("-" * 40)
print(df.isnull().sum())

print("\nData Types:")
print("-" * 40)
print(df.dtypes)
```

```python
# ================================================
# CELL 9: Fix Data Types & Add Useful Columns
# ================================================

# Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract useful columns from Order Date
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%b')

# Calculate shipping duration (days taken to ship)
df['Ship Duration'] = (df['Ship Date'] - df['Order Date']).dt.days

print("✅ Date columns fixed!")
print("\nNew columns added:")
print(df[['Order Date', 'Year', 'Month', 'Month Name', 'Ship Duration']].head())
print("\nYears in dataset:", sorted(df['Year'].unique().tolist()))
```

## 📊 Section 3: Exploratory Data Analysis (EDA)

Analyzing sales trends, profits, customer segments,
regional performance, and product categories.

### Charts we will create:
1. Yearly Sales Trend
2. Monthly Sales Pattern
3. Sales by Category
4. Sales & Profit by Region
5. Top 10 Products by Sales
6. Customer Segment Analysis
7. Discount vs Profit Relationship

```python
# ================================================
# CELL 11: Yearly Sales Trend
# ================================================

yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

plt.figure(figsize=(9, 5))
plt.plot(yearly_sales['Year'], yearly_sales['Sales'], 
         marker='o', linewidth=2.5, 
         color='#2196F3', markersize=8)

for i, row in yearly_sales.iterrows():
    plt.text(row['Year'], row['Sales'] + 8000, 
             f"${row['Sales']:,.0f}", 
             ha='center', fontweight='bold')

plt.title('Yearly Sales Trend (2014–2017)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.savefig('images/yearly_sales_trend.png', dpi=150)
plt.show()

print("Yearly Sales:")
print("-" * 40)
print(yearly_sales)
```

```python
# ================================================
# CELL 12: Sales by Category
# ================================================

category_sales = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
category_sales = category_sales.sort_values('Sales', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sales by Category
axes[0].bar(category_sales['Category'], category_sales['Sales'],
            color=['#2196F3', '#4CAF50', '#FF9800'])
axes[0].set_title('Total Sales by Category', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Total Sales ($)')
for i, v in enumerate(category_sales['Sales']):
    axes[0].text(i, v + 5000, f"${v:,.0f}", ha='center', fontweight='bold')

# Profit by Category
axes[1].bar(category_sales['Category'], category_sales['Profit'],
            color=['#2196F3', '#4CAF50', '#FF9800'])
axes[1].set_title('Total Profit by Category', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Total Profit ($)')
for i, v in enumerate(category_sales['Profit']):
    axes[1].text(i, v + 1000, f"${v:,.0f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('images/sales_profit_by_category.png', dpi=150)
plt.show()

print("Sales & Profit by Category:")
print("-" * 40)
print(category_sales)
```

```python
# ================================================
# CELL 13: Sales & Profit by Region
# ================================================

region_data = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
region_data = region_data.sort_values('Sales', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

colors = ['#2196F3', '#4CAF50', '#FF9800', '#E91E63']

# Sales by Region
axes[0].bar(region_data['Region'], region_data['Sales'], color=colors)
axes[0].set_title('Total Sales by Region', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Total Sales ($)')
for i, v in enumerate(region_data['Sales']):
    axes[0].text(i, v + 3000, f"${v:,.0f}", ha='center', fontweight='bold', fontsize=9)

# Profit by Region
axes[1].bar(region_data['Region'], region_data['Profit'], color=colors)
axes[1].set_title('Total Profit by Region', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Total Profit ($)')
for i, v in enumerate(region_data['Profit']):
    axes[1].text(i, v + 500, f"${v:,.0f}", ha='center', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.savefig('images/sales_profit_by_region.png', dpi=150)
plt.show()

print("Sales & Profit by Region:")
print("-" * 40)
print(region_data)
```

```python
# ================================================
# CELL 14: Top 10 Products by Sales
# ================================================

top_products = df.groupby('Product Name')['Sales'].sum()\
                 .sort_values(ascending=False)\
                 .head(10)\
                 .reset_index()

plt.figure(figsize=(12, 6))
bars = plt.barh(top_products['Product Name'], 
                top_products['Sales'],
                color='#2196F3')

plt.title('Top 10 Products by Sales', fontsize=14, fontweight='bold')
plt.xlabel('Total Sales ($)')
plt.gca().invert_yaxis()

for bar, v in zip(bars, top_products['Sales']):
    plt.text(bar.get_width() + 200, bar.get_y() + bar.get_height()/2,
             f"${v:,.0f}", va='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('images/top10_products.png', dpi=150)
plt.show()

print("Top 10 Products by Sales:")
print("-" * 40)
print(top_products)
```

```python
# ================================================
# CELL 15: Customer Segment Analysis
# ================================================

segment_data = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()
segment_data = segment_data.sort_values('Sales', ascending=False)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Sales by Segment
axes[0].pie(segment_data['Sales'],
            labels=segment_data['Segment'],
            autopct='%1.1f%%',
            colors=['#2196F3', '#4CAF50', '#FF9800'],
            startangle=90)
axes[0].set_title('Sales by Segment', fontsize=13, fontweight='bold')

# Profit by Segment
axes[1].pie(segment_data['Profit'],
            labels=segment_data['Segment'],
            autopct='%1.1f%%',
            colors=['#2196F3', '#4CAF50', '#FF9800'],
            startangle=90)
axes[1].set_title('Profit by Segment', fontsize=13, fontweight='bold')

# Order Count by Segment
segment_orders = df.groupby('Segment')['Order ID'].count().reset_index()
axes[2].bar(segment_orders['Segment'], segment_orders['Order ID'],
            color=['#2196F3', '#4CAF50', '#FF9800'])
axes[2].set_title('Order Count by Segment', fontsize=13, fontweight='bold')
axes[2].set_xlabel('Segment')
axes[2].set_ylabel('Number of Orders')

plt.tight_layout()
plt.savefig('images/segment_analysis.png', dpi=150)
plt.show()

print("Segment Analysis:")
print("-" * 40)
print(segment_data)
```

```python
# ================================================
# CELL 16: Discount vs Profit Relationship
# ================================================

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot: Discount vs Profit
axes[0].scatter(df['Discount'], df['Profit'],
                alpha=0.4, color='#2196F3', edgecolors='none')
axes[0].axhline(y=0, color='red', linestyle='--', linewidth=1.5)
axes[0].set_title('Discount vs Profit', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Discount')
axes[0].set_ylabel('Profit ($)')

# Average profit by discount level
discount_profit = df.groupby('Discount')['Profit'].mean().reset_index()
axes[1].bar(discount_profit['Discount'].astype(str), 
            discount_profit['Profit'],
            color=discount_profit['Profit'].apply(
                lambda x: '#4CAF50' if x > 0 else '#F44336'))
axes[1].axhline(y=0, color='black', linestyle='--', linewidth=1)
axes[1].set_title('Average Profit by Discount Level', 
                  fontsize=13, fontweight='bold')
axes[1].set_xlabel('Discount')
axes[1].set_ylabel('Average Profit ($)')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('images/discount_vs_profit.png', dpi=150)
plt.show()

# Key finding
high_discount = df[df['Discount'] >= 0.5]
print(f"Orders with 50%+ discount: {len(high_discount)}")
print(f"Average profit at 50%+ discount: ${high_discount['Profit'].mean():.2f}")
print(f"Total loss from high discounts: ${high_discount['Profit'].sum():.2f}")
```

```python
# ================================================
# CELL 17: Monthly Sales Pattern
# ================================================

monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

plt.figure(figsize=(14, 6))

colors = ['#2196F3', '#4CAF50', '#FF9800', '#E91E63']
years = sorted(df['Year'].unique())

for i, year in enumerate(years):
    yearly_data = monthly_sales[monthly_sales['Year'] == year]
    plt.plot(yearly_data['Month'], yearly_data['Sales'],
             marker='o', label=str(year),
             color=colors[i], linewidth=2)

plt.title('Monthly Sales Pattern by Year', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(range(1, 13), 
           ['Jan','Feb','Mar','Apr','May','Jun',
            'Jul','Aug','Sep','Oct','Nov','Dec'])
plt.legend(title='Year')
plt.tight_layout()
plt.savefig('images/monthly_sales_pattern.png', dpi=150)
plt.show()

print("Best performing months overall:")
print("-" * 40)
best_months = df.groupby('Month')['Sales'].sum()\
                .sort_values(ascending=False).head(3)
print(best_months)
```

## 💡 Section 4: Key Insights & Conclusions

Summary of the most important business findings from this analysis.

```python
# ================================================
# CELL 19: Final Insights Summary
# ================================================

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_sales) * 100

print("=" * 60)
print("     SUPERSTORE SALES ANALYSIS — KEY INSIGHTS")
print("=" * 60)
print(f"""
📊 BUSINESS OVERVIEW
→ Total Sales         : ${total_sales:,.2f}
→ Total Profit        : ${total_profit:,.2f}
→ Overall Profit Margin: {profit_margin:.1f}%
→ Total Orders        : {df['Order ID'].nunique()}
→ Total Customers     : {df['Customer ID'].nunique()}
→ Period              : 2014 – 2017

📈 SALES TRENDS
→ Sales grew 51% from 2014 ($484K) to 2017 ($733K)
→ Best months: November, December, September (Q4 peak)
→ West region leads with $725K in sales

🏷️ CATEGORY INSIGHTS
→ Technology = highest sales ($836K) and profit ($145K)
→ Furniture   = high sales ($741K) but very low profit ($18K)
→ Furniture profit margin is critically low — needs review

👥 CUSTOMER SEGMENTS
→ Consumer segment = 50%+ of total sales
→ All 3 segments are profitable

⚠️ CRITICAL FINDING — DISCOUNT IMPACT
→ 922 orders had 50%+ discount
→ Average profit at high discount = -$105 per order
→ Total loss from high discounts = -$97,065
→ Reducing excessive discounts could recover $97K in profit

💡 BUSINESS RECOMMENDATIONS
1. Review Furniture pricing — high sales, very low profit
2. Reduce discounts above 50% — causing $97K in losses
3. Invest more in Technology — best profit margin
4. Focus Q4 marketing — November & December peak months
5. Investigate Central region — low profit despite decent sales
""")
print("=" * 60)
```
